```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

initial_data = pd.read_csv("maaslar_yeni.csv")
```

```python
initial_data.head()
```

```python
initial_data.describe()
```

```python
x_df = initial_data.iloc[:,2:5]
y_df = initial_data.iloc[:,-1:]
```

```python
x_df
```

```python
from sklearn import preprocessing
ohe = preprocessing.OneHotEncoder()
```

```python
unvan_column = x_df.iloc[:,0:1].values
unvan_column
```

```python
unvan_column = ohe.fit_transform(unvan_column).toarray()
```

```python
unvan_column
```

```python
unvan_df = pd.DataFrame(data=unvan_column, index=range(len(unvan_column)), columns=["Cayci","Sekreter","Uzman Yardimcisi","Uzman","Proje Yoneticisi","Sef","Mudur","Direktor","C-level","CEO"])
unvan_df
```

```python
x_df.drop(columns=["UnvanSeviyesi"], inplace=True)
x_df = pd.concat([x_df,unvan_df], axis=1)
x_df
```

```python
y_df
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_df,y_df,test_size=0.25,random_state=0)
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
print(x_test.shape,x_train.shape)
```

```python
X_test = sc.fit_transform(x_test.iloc[:,0:2])
X_train = sc.fit_transform(x_train.iloc[:,0:2])
```

```python
print(X_test.shape,X_train.shape)
```

```python
X_test = np.concatenate((X_test,x_test.iloc[:,2:]), axis=1)
X_train = np.concatenate((X_train,x_train.iloc[:,2:]), axis=1)
```

#### Multi Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)

lr_preds = lr.predict(X_test)
lr_preds
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(lr_preds,y_test)
```

#### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

pol_reg2 = PolynomialFeatures(degree=4)
X_poly = pol_reg2.fit_transform(X_train)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y_train)
```

```python
polyreg_preds = lin_reg2.predict(pol_reg2.fit_transform(X_test))

r2_score(polyreg_preds, y_test)
```

```python
Y_train = sc.fit_transform(y_train)
Y_test = sc.fit_transform(y_test)
```

```python
Y_train.ravel()
```

#### Support Vector Machine Regression

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel = 'rbf')

svr_reg.fit(X_train, Y_train.ravel())
svr_preds = svr_reg.predict(X_test)
```

```python
svr_preds
```

```python
r2_score(svr_preds, Y_test.ravel())
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
```

```python
dtr.fit(x_train, y_train)

r2_score(dtr.predict(x_test),y_test)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators = 10,random_state=0)
rf_reg.fit(x_train,y_train.values.ravel())
```

```python
r2_score(rf_reg.predict(x_test),y_test)
```
