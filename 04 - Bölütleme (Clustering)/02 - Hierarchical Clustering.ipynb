<h1 style="text-align:center;">Hierarchical Clustering</h1>

İki farklı yaklaşımla çalışır:

- **Agglomerative:**
    - Buttom-up approach.
    - Her veri tek bir küme/bölüt ile başlar.
    - En yakın ikişer komşuyu alıp ikişerli küme/bölüt oluşturulur.
    - En yakın iki kümeyi alıp yeni bir bölüt oluşturur.
    - Bir önceki adım, tek bir küme/bölüt olana kadar devam eder.
- **Divisive:**
    - Top-down approach.

En iyi bölme noktası en uzun bölme aralığının bulunduğu noktadır. Bu örnekte 4, $(2, 5)$ ile $((2, 5), 4)$ arası.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("musteriler.csv")
data.head()
```

```python
X = data.iloc[:,3:].values
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
Y_tahmin = ac.fit_predict(X)
```

```python
plt.scatter(X[Y_tahmin==0,0], X[Y_tahmin==0,1], s=100, c="red")
plt.scatter(X[Y_tahmin==1,0], X[Y_tahmin==1,1], s=100, c="blue")
plt.scatter(X[Y_tahmin==2,0], X[Y_tahmin==2,1], s=100, c="green")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method="ward"))
plt.show()
```
